# BLOOM FILTER EXP

```python
config1 = "5 hashes, 10^7  bits"
generation_time1 = [ 0.075737, 0.301653, 2.975041, 33.862452 ]
read_time_full1 =  [ 0.059771, 0.560084, 2.812387, 5.691733 ]
read_time_half1 = [ 0.027026, 0.285551, 1.372883, 2.738212 ]
updates_time1 = [ 0.038411, 0.397354, 2.111856, 4.311012, 21.150731 ]
```

```python
config2 = "10 hashes, 10^7  bits"
generation_time2 = [ 0.076804, 0.347260, 3.669012, 40.347595 ]
read_time_full2 = [ 0.056622, 0.591920, 2.978903, 6.014575 ]
read_time_half2 = [ 0.028577, 0.301173, 1.428047, 2.890665 ]
updates_time2 = [ 0.043090, 0.460819, 2.471898, 5.134752, 24.794146 ]
```

```python
config3 = "5 hashes, 10^6  bits"
generation_time3 = [ 0.031942, 0.263881, 2.920011, 31.294598 ]
read_time_full3 = [ 0.054419, 0.557745, 2.879704, 5.879081 ]
read_time_half3 =  [ 0.038659, 0.405856, 2.116863, 4.868357 ]
updates_time3 = [ 0.042532, 0.435993, 2.187626, 4.461553, 21.739211 ]
```

```python
config4 = "10 hashes, 10^6  bits"
generation_time4 = [ 0.035020, 0.297071, 3.719979, 35.599067 ]
read_time_full4 = [ 0.060737, 0.601402, 3.078153, 6.278124 ]
read_time_half4 =  [ 0.041370, 0.435714, 2.241637, 4.423537 ]
updates_time4 = [ 0.039975, 0.422394, 2.203160, 4.605589, 22.728689 ]
```

```python
# Generation
val_generation = [10000, 100000, 1000000, 10000000]
plt.figure(figsize=(12,8))
plt.plot(val_generation, generation_time1, color = 'red', label = config1)
plt.plot(val_generation, generation_time2, color = 'green', label = config2)
plt.plot(val_generation, generation_time3, color = 'yellow', label = config3)
plt.plot(val_generation, generation_time4, color = 'blue', label = config4)
plt.xlabel('Number of keys inserted')
plt.ylabel('Execution Time')
plt.title("Execution Time on LSM tree generation with Bloom filter")
plt.legend(loc = 1)
plt.xscale('log')
plt.show()
```

```python
# Reading
val_reads = [1000, 10000, 50000, 100000]
plt.figure(figsize=(12,8))
plt.plot(val_reads, read_time_full1, linestyle='--', color = 'red', label = 'full ' + config1)
plt.plot(val_reads, read_time_half1, color = 'red', label = 'half ' + config1)
plt.plot(val_reads, read_time_full2, linestyle='--', color = 'green', label = 'full ' + config2)
plt.plot(val_reads, read_time_half2, color = 'green', label = 'half ' + config2)
plt.plot(val_reads, read_time_full3, linestyle='--', color = 'yellow', label = 'full ' + config3)
plt.plot(val_reads, read_time_half3, color = 'yellow', label = 'half ' + config3)
plt.plot(val_reads, read_time_full4, linestyle='--', color = 'blue', label = 'full ' + config4)
plt.plot(val_reads, read_time_half4, color = 'blue', label = 'half ' + config4)
plt.xlabel('Number of reads')
plt.ylabel('Execution Time')
plt.title("Execution Time on batch reading with Bloom filter")
plt.legend(loc = 1)
plt.xscale('log')
plt.show()
```

```python
# Updates
val_updates = [10000, 100000, 500000, 1000000, 5000000]
plt.figure(figsize=(12,8))
plt.plot(val_updates, updates_time1, color = 'red', label = 'half ' + config1)
plt.plot(val_updates, updates_time2, color = 'green', label = 'half ' + config2)
plt.plot(val_updates, updates_time3, color = 'yellow', label = 'half ' + config3)
plt.plot(val_updates, updates_time4, color = 'blue', label = 'half ' + config4)
plt.xlabel('Number of updates')
plt.ylabel('Execution Time')
plt.title("Execution Time on batch updates with Bloom filter")
plt.legend(loc = 1)
plt.xscale('log')
plt.show()
```

# Architecture EXP

```python
generation_time = [ 3.417755, 3.774972, 3.479524, 3.065683, 3.135843, 3.086700 ]
read_uniform_time = [ 0.581283, 0.579361, 0.551461, 0.696590, 0.433572, 0.557987 ]
read_skewed_beginning_time = [ 0.548716, 0.600256, 0.600505, 0.567601, 0.430630, 0.536001 ]
read_skewed_end_time = [ 0.528562, 0.547915, 0.617458, 0.535801, 0.449455, 0.548006 ]
udpate_time = [ 0.604489, 0.484911, 0.552530, 0.448643, 0.416557, 0.548409 ]
update_skewed_beginning_time = [ 0.592825, 0.371829, 0.488737, 0.403457, 0.395107, 0.560208 ]
update_skewed_end_time = [ 0.708643, 0.365318, 0.538887, 0.388185, 0.394662, 0.559185 ]
```

```python
# We want the results in throughput
num_elements = 1000000;
num_reads = 10000;
num_updates = 100000;

generation_throughput = num_elements / np.array(generation_time)
read_uniform_throughput = num_reads / np.array(read_uniform_time)
read_skewed_beginning_throughput = num_reads / np.array(read_skewed_beginning_time)
read_skewed_end_throughput = num_reads / np.array(read_skewed_end_time)
udpate_throughput = num_updates / np.array(udpate_time)
update_skewed_beginning_throughput = num_updates / np.array(update_skewed_beginning_time)
update_skewed_end_throughput = num_updates / np.array(update_skewed_end_time)
```

```python
# Bar plots

fig = plt.figure()
ax = fig.add_subplot(111)

ind = np.arange(len(generation_time))
width = 0.5   

generation = ax.bar(ind, generation_throughput, width,
                color='blue')

ax.set_xlim(-width,len(ind)+width)
#ax.set_ylim(0,45)
ax.set_ylabel('Throughput (keys/s)')
ax.set_title('Throughput on generation for different architectures')
xTickMarks = ['LSMT '+str(i+1) for i in range(len(generation_time))]
ax.set_xticks(ind+width)
xtickNames = ax.set_xticklabels(xTickMarks)
plt.setp(xtickNames, rotation=45, fontsize=10)

## add a legend
#ax.legend( (generation[0]), ('Uniform reads', 'Skewed reads on beginning', 'Skewed reads on end'), loc=2 )

plt.show()
```

```python
# Bar plots

fig = plt.figure()
ax = fig.add_subplot(111)

ind = np.arange(len(generation_time))
width = 0.20    

reads_u = ax.bar(ind, read_uniform_throughput, width,
                color='blue')

reads_sb = ax.bar(ind+width, read_skewed_beginning_throughput, width,
                    color='red')

reads_se = ax.bar(ind+2*width, read_skewed_end_throughput, width,
                color='green')

ax.set_xlim(-width,len(ind)+width)
#ax.set_ylim(0,45)
ax.set_ylabel('Throughput (keys/s)')
ax.set_title('Throughput on reads for different architectures')
xTickMarks = ['LSMT '+str(i+1) for i in range(len(generation_time))]
ax.set_xticks(ind+width)
xtickNames = ax.set_xticklabels(xTickMarks)
plt.setp(xtickNames, rotation=45, fontsize=10)

## add a legend
ax.legend( (reads_u[0], reads_sb[0], reads_se[0]), ('Uniform reads', 'Skewed reads on beginning', 'Skewed reads on end'), loc=2 )

plt.show()
```

```python
# Bar plots

fig = plt.figure()
ax = fig.add_subplot(111)

ind = np.arange(len(generation_time))
width = 0.20    

updates_u = ax.bar(ind, udpate_throughput, width,
                color='blue')

updates_sb = ax.bar(ind+width, update_skewed_beginning_throughput, width,
                    color='red')

updates_se = ax.bar(ind+2*width, update_skewed_end_throughput, width,
                color='green')

ax.set_xlim(-width,len(ind)+width)
ax.set_ylim(0,350000)
ax.set_ylabel('Throughput (keys/s)')
ax.set_title('Throughput on updates for different architectures')
xTickMarks = ['LSMT '+str(i+1) for i in range(len(generation_time))]
ax.set_xticks(ind+width)
xtickNames = ax.set_xticklabels(xTickMarks)
plt.setp(xtickNames, rotation=45, fontsize=10)

## add a legend
ax.legend( (updates_u[0], updates_sb[0], updates_se[0]), ('Uniform updates', 'Skewed updates on beginning', 'Skewed updates on end'), loc=1 )

plt.show()
```

# BOUNDS CHECK

```python
read_uniform_time = [4.198531, 5.703282, 4.978859, 4.843169]
read_skewed_beginning_time = [4.380644, 5.488670, 5.341354, 4.956924]
read_skewed_middle_time = [4.386300, 5.546905, 5.284514, 4.766002]
read_skewed_end_time = [3.322344, 5.475629, 3.390869, 4.643080]
```

```python
# Throughput
read_uniform_throughput = 10**6 / np.array(read_uniform_time)
read_skewed_beginning_throughput = 10**6 / np.array(read_skewed_beginning_time)
read_skewed_middle_throughput = 10**6 / np.array(read_skewed_middle_time)
read_skewed_end_throughput = 10**6 / np.array(read_skewed_end_time)
```

```python
# Bar plots

fig = plt.figure()
ax = fig.add_subplot(111)

ind = np.arange(len(read_uniform_throughput))
width = 0.20    

reads_u = ax.bar(ind, read_uniform_throughput, width,
                color='blue')

reads_sb = ax.bar(ind+width, read_skewed_beginning_throughput, width,
                    color='red')
reads_sm = ax.bar(ind+2*width, read_skewed_middle_throughput, width,
                    color='gray')

reads_se = ax.bar(ind+3*width, read_skewed_end_throughput, width,
                color='green')

ax.set_xlim(-width,len(ind)+width)
#ax.set_ylim(0,350000)
ax.set_ylabel('Throughput (keys/s)')
ax.set_title('Throughput on reads with/without Bounds Check')
xTickMarks = ['Sorted with BC', 'Unsorted with BC', 'Sorted without BC', 'Unsorted without BC']
ax.set_xticks(ind+width)
xtickNames = ax.set_xticklabels(xTickMarks)
plt.setp(xtickNames, rotation=45, fontsize=10)

## add a legend
ax.legend( (reads_u[0], reads_sb[0], reads_sm[0], reads_se[0]), ('Uniform', 'Skewed on beginning', 'Skewed on middle', 'Skewed on end'), loc=9 )

plt.show()
```
